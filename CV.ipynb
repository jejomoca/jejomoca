```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
from sklearn.linear_model import LinearRegression

years = np.arange(2015, 2025)  
prices_basic = [7.99, 7.99, 7.99, 7.99, 7.99, np.nan, np.nan, np.nan, np.nan, np.nan]  # Plan básico se elimina en 2023
prices_standard = [9.99, 9.99, 10.99, 10.99, 11.99, 11.99, 12.99, 12.99, 12.99, 12.99]  # Cambios registrados
prices_premium = [11.99, 11.99, 13.99, 13.99, 15.99, 15.99, 17.99, 17.99, 17.99, 17.99]  # Cambios registrados
prices_ad_supported = [np.nan, np.nan, np.nan, np.nan, np.nan, 5.49, 5.49, 5.49, 5.49, 5.49]  # Introducido en 2022

average_prices = []
for i in range(len(years)):
    prices = [prices_basic[i], prices_standard[i], prices_premium[i]]
    average_prices.append(np.nanmean(prices))  


subscribers = [70.5, 79.9, 100.1, 150.3, 180.5, 190.7, 210.9, 230.1, 260.3, 290.5]


model = LinearRegression()
model.fit(years.reshape(-1, 1), subscribers)
future_years = np.array([2025, 2026])
predicted_subscribers = model.predict(future_years.reshape(-1, 1))


fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.bar(years, average_prices, color='blue', label='Precio Promedio Sin Anuncios (€)', align='center', alpha=0.7)
ax1.bar(years, prices_ad_supported, bottom=average_prices, color='red', label='Precio Plan con Anuncios (€)', align='center', alpha=0.9)


ax1.set_xlabel('Año')
ax1.set_ylabel('Precio (€)')
ax1.set_title('Evolución de los Precios y Suscriptores de Netflix en España')
ax1.set_ylim(0, 25)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(np.concatenate((years, future_years)), np.concatenate((subscribers, predicted_subscribers)), 'o--', color='black', label='Suscriptores (millones)')
ax2.set_ylabel('Suscriptores (millones)')
ax2.set_ylim(0, 400)
ax2.legend(loc='upper right')

ax1.set_xticks(np.concatenate((years, future_years)))
ax1.set_xticklabels(np.concatenate((years, future_years)))
ax1.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{int(x)}'))

plt.grid(True, linestyle='--', which='both', axis='y', alpha=0.9)
plt.show()

```
